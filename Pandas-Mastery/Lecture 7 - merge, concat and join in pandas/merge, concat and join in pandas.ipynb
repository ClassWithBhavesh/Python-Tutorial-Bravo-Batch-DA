### ***Merging |concat | joining in Pandas***

```python
import pandas as pd
```

```python
ds1 = pd.Series(data=['Z', 'X', 'Z', 'Y', 'X', 'Z', 'Y'])
ds2 = pd.Series(data=[10,20,30,40,50,60,70])

ds3 = pd.Series(['Z', 'X', 'Z', 'Y'])
ds4 = pd.Series([23,27,34,41])

df1 = pd.DataFrame({
    'axis': ds1,
    'quantity': ds2
})

df2 = pd.DataFrame({
    'axis': ds3,
    'age': ds4
})

# Joining in pandas
joinedDf = df1.join(df2.set_index('axis'), on='axis')
print(joinedDf)


# Merging in Pandas -
# mergedDf = pd.merge(df2, df1, on='axis')
# print(mergedDf)
print(df1)
print(df2)

# concating in Pandas -
# concatedDf = pd.concat([df1, df2], axis=1)
# print(concatedDf)
# if axis - 1 --> column concatenation ---> all the columns must be unique, if two or more columns are same than it will scattered the whole data
# if axis - 0 --> row concatenation    ---> there must be at least one same column


# li = [[10,20,30], [40,50,60], [70,80,90]]

# df = pd.DataFrame(li, columns=('x-cors', 'y-cors', 'z-cors'), index=['Aa', 'Bb', 'Cc'])
# print(df)
```
